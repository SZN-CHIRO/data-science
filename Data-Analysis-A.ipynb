```python
# Step 1: Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load a built-in dataset from Seaborn (environment/weather style)
df = sns.load_dataset("diamonds")  # Replaceable with other datasets like 'penguins', 'titanic', etc.

# Step 3: Explore the data
print(df.head())
print(df.info())
print(df.describe())

# Step 4: Visualize the distribution of carat (as proxy for quantity/size)
plt.figure(figsize=(10, 5))
sns.histplot(df['carat'], bins=30, kde=True, color='green')
plt.title("Distribution of Diamond Carat")
plt.xlabel("Carat")
plt.ylabel("Frequency")
plt.show()

# Step 5: Boxplot - Price by Cut Quality
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='cut', y='price', palette="Set2")
plt.title("Diamond Price by Cut Quality")
plt.show()

# Step 6: Scatterplot - Carat vs Price
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='carat', y='price', hue='cut', alpha=0.6)
plt.title("Carat vs Price Colored by Cut")
plt.show()

# Step 7: Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
